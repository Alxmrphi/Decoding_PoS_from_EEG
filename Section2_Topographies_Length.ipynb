## Plan

* Load all .FIF files containing EEG data
* Take grand average ERP
* Plot topography at multiple points
* Save files
* Organise plots into ICA-specific sections
* Under "Strong ICA" then "epoch" and 3 plots, len, freq, class
* Under "epoch" is "sentence bc" and under that, "no bc"
* Then repeat for "Weak ICA" and "No ICA"

NB: need to apply baseline correction on the .fif files that aren't labelled "bc"

```python
import numpy as np
import mne
import os
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['svg.fonttype'] = 'none'
matplotlib.rcParams['font.family'] = 'Arial'
```

```python
data_folder = './data_visualisation'
```

# Length data

## ICA strong

```python
# ica strong epoch bc
ica_strong_epoch_bc = mne.read_epochs(os.path.join(data_folder, 'len_matched_epochs_ica_strong_epoch_bc_train-epo.fif'))
ica_strong_epoch_bc_erp = ica_strong_epoch_bc.average()
ica_strong_epoch_bc_erp_cropped = ica_strong_epoch_bc_erp.crop(0,0.55)
del ica_strong_epoch_bc
```

```python
# ica strong sent bc
ica_strong_sent_bc = mne.read_epochs(os.path.join(data_folder, 'len_matched_epochs_ica_strong_sent_bc_train-epo.fif'))
ica_strong_sent_bc_erp = ica_strong_sent_bc.average()
ica_strong_sent_bc_erp_cropped = ica_strong_sent_bc_erp.crop(0,0.55)
del ica_strong_sent_bc
```

```python
# ica strong no bc
ica_strong_no_bc = mne.read_epochs(os.path.join(data_folder, 'len_matched_epochs_ica_strong_no_bc_train-epo.fif'))
ica_strong_no_bc_erp = ica_strong_no_bc.average()
ica_strong_no_bc_erp_cropped = ica_strong_no_bc_erp.crop(0,0.55)
del ica_strong_no_bc
```
